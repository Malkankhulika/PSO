## **1. Sektor barang konsumen primer: Saham Unilever**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca dataset
data = pd.read_csv('/content/SAHAM UNLIVER INDONESIA.csv')
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data = data.sort_values(by='Date')

# Harga Penutupan dan Tanggal
close_prices = data['Close'].values
dates = data['Date']

# Menghitung log return historis
log_returns = np.log(close_prices[1:] / close_prices[:-1])

# Skenario 1: Simulasi Random Walk
np.random.seed(42)
simulated_returns_rw = np.random.normal(log_returns.mean(), log_returns.std(), len(log_returns))
simulated_prices_rw = [close_prices[0]]

for ret in simulated_returns_rw:
    simulated_prices_rw.append(simulated_prices_rw[-1] * np.exp(ret))

# Skenario 2: Tren Historis Bullish (Kenaikan Stabil)
bullish_trend = log_returns.mean() + log_returns.std() * 0.5
simulated_prices_bullish = [close_prices[0]]

for _ in range(len(log_returns)):
    ret = np.random.normal(bullish_trend, log_returns.std())
    simulated_prices_bullish.append(simulated_prices_bullish[-1] * np.exp(ret))

# Skenario 3: Tren Historis Bearish (Penurunan Stabil)
bearish_trend = log_returns.mean() - log_returns.std() * 0.5
simulated_prices_bearish = [close_prices[0]]

for _ in range(len(log_returns)):
    ret = np.random.normal(bearish_trend, log_returns.std())
    simulated_prices_bearish.append(simulated_prices_bearish[-1] * np.exp(ret))

# Plot hasil simulasi
plt.figure(figsize=(16, 8))
plt.plot(dates, close_prices, label='Harga Asli (Saham Uniliver Indonesia)', color='black', alpha=0.8)
plt.plot(dates[:len(simulated_prices_rw)], simulated_prices_rw, label='Simulasi Random Walk', color='blue', linestyle='--', alpha=0.8)
plt.plot(dates[:len(simulated_prices_bullish)], simulated_prices_bullish, label='Skenario Bullish', color='orange', linestyle='-.', alpha=0.8)
plt.plot(dates[:len(simulated_prices_bearish)], simulated_prices_bearish, label='Skenario Bearish', color='green', linestyle=':', alpha=0.8)

# Mengatur batas sumbu Y
plt.ylim(0, 4000)  # Mengatur nilai minimum sumbu Y ke 0
# Anda juga dapat mengatur batas maksimum, misalnya:
# plt.ylim(0, 10000)  # Sesuaikan nilai maksimum sesuai kebutuhan

plt.title('Simulasi Model Random Walk dan Tren Historis pada Harga Saham Uniliver Indonesia')
plt.xlabel('Bulan')
plt.ylabel('Harga')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

# **Penjelasan**

**1. Harga Asli (Garis Hitam)**


*   Pola pergerakan harga yang menunjukkan adanya tren jangka panjang dan volatilitas pasar.
*   Harga saham Unilever Indonesia mengalami fluktuasi tajam pada bulan September hingga Oktober, menunjukkan bahwa faktor eksternal (seperti berita atau kondisi pasar) mungkin sangat mempengaruhi harga saham pada periode tersebut.
* Terdapat penurunan yang jelas dalam harga saham, terutama dari awal tahun 2024 hingga pertengahan tahun 2024, di mana harga saham mencapai titik terendah.

**2. Simulasi Random Walk (Garis Biru)**
*   Hasil dari model random walk yang cenderung lebih acak dan tidak menunjukkan tren yang jelas, terutama pada akhir tahun.
*   Fluktuasi harga pada simulasi lebih rendah, mencerminkan bahwa pergerakan harga acak tidak menangkap pola besar yang ada dalam data asli. Ini mencerminkan ketidakpastian yang sering terjadi di pasar saham.

**3. Tren Bullish**
*   Garis kuning menunjukkan skenario bullish, di mana harga saham diperkirakan akan meningkat secara stabil.
*   Namun, dalam grafik ini, skenario bullish tidak terlihat realistis karena harga asli menunjukkan tren penurunan. Ini menunjukkan bahwa meskipun ada potensi untuk kenaikan, kondisi pasar saat ini tidak mendukung skenario tersebut.

**4. Tren Bearish**
* Garis hijau putus-putus menunjukkan skenario bearish, di mana harga saham diperkirakan akan menurun secara stabil.
* Skenario ini lebih sesuai dengan tren harga asli yang terlihat, terutama pada periode penurunan yang signifikan. Ini menunjukkan bahwa jika tren saat ini berlanjut, harga saham mungkin akan terus menurun.
