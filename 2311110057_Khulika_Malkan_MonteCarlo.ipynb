```python
# halaman 27 (output)

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Fungsi untuk menghitung integral
def f1(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

def f2(x):
    return 3 * x**2

# Menghitung integral eksak
def exact_integral(func, a, b):
    if func == f1:
        return quad(f1, a, b)[0]  # Hasil integral eksak untuk f1
    elif func == f2:
        return (b**3 - a**3)  # Hasil integral eksak untuk f2

# Metode Rectangular
def rectangular_integral(func, a, b, n):
    x = np.linspace(a, b, n)
    dx = (b - a) / n
    return np.sum(func(x[:-1])) * dx

# Metode Trapezoidal
def trapezoidal_integral(func, a, b, n):
    x = np.linspace(a, b, n)
    dx = (b - a) / (n - 1)
    return (dx / 2) * (func(a) + 2 * np.sum(func(x[1:-1])) + func(b))

# Fungsi Monte Carlo
def monte_carlo_integral(func, a, b, n):
    x_random = np.random.uniform(a, b, n)
    return (b - a) * np.mean(func(x_random))

# Parameter
a1, b1 = -5, 5          # Batas untuk f1
a2, b2 = -5, 5          # Batas untuk f2
n = 10000

# Hasil untuk fungsi pertama
exact_f1 = exact_integral(f1, a1, b1)
rectangular_f1 = rectangular_integral(f1, a1, b1, n)
trapezoidal_f1 = trapezoidal_integral(f1, a1, b1, n)
monte_carlo_f1 = monte_carlo_integral(f1, a1, b1, n)

# Hasil untuk fungsi kedua
exact_f2 = exact_integral(f2, a2, b2)
rectangular_f2 = rectangular_integral(f2, a2, b2, n)
trapezoidal_f2 = trapezoidal_integral(f2, a2, b2, n)
monte_carlo_f2 = monte_carlo_integral(f2, a2, b2, n)

# Print hasil
print("Fungsi 1: (1 / √(2π)) * exp(-x² / 2)")
print(f"Integral Eksak: {exact_f1:.4f}")
print(f"Integral Rectangular: {rectangular_f1:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f1:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f1:.4f}")

print("\nFungsi 2: 3x²")
print(f"Integral Eksak: {exact_f2:.4f}")
print(f"Integral Rectangular: {rectangular_f2:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f2:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f2:.4f}")

# Visualisasi fungsi
x1 = np.linspace(a1, b1, 100)
x2 = np.linspace(a2, b2, 100)

plt.figure(figsize=(12, 6))

# Subplot untuk fungsi pertama
plt.subplot(1, 2, 1)
plt.plot(x1, f1(x1), color='blue', label='(1 / √(2π)) * exp(-x² / 2)')
plt.fill_between(x1, f1(x1), color='lightblue', alpha=0.5)
plt.title('Fungsi 1: (1 / √(2π)) * exp(-x² / 2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Subplot untuk fungsi kedua
plt.subplot(1, 2, 2)
plt.plot(x2, f2(x2), color='orange', label='3x²')
plt.fill_between(x2, f2(x2), color='lightyellow', alpha=0.5)
plt.title('Fungsi 2: 3x²')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Fungsi untuk menghitung integral
def f1(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

def f2(x):
    return 3 * x**2

def f3(x):
    return np.sin(x)

def f4(x):
    return np.exp(-x**2)

# Menghitung integral eksak
def exact_integral(func, a, b):
    if func == f1:
        return quad(f1, a, b)[0]
    elif func == f2:
        return (b**3 - a**3)
    elif func == f3:
        return quad(f3, a, b)[0]
    elif func == f4:
        return quad(f4, a, b)[0]

# Metode Rectangular
def rectangular_integral(func, a, b, n):
    x = np.linspace(a, b, n)
    dx = (b - a) / n
    return np.sum(func(x[:-1])) * dx

# Metode Trapezoidal
def trapezoidal_integral(func, a, b, n):
    x = np.linspace(a, b, n)
    dx = (b - a) / (n - 1)
    return (dx / 2) * (func(a) + 2 * np.sum(func(x[1:-1])) + func(b))

# Fungsi Monte Carlo
def monte_carlo_integral(func, a, b, n):
    x_random = np.random.uniform(a, b, n)
    return (b - a) * np.mean(func(x_random))

# Parameter untuk fungsi pertama
a1, b1 = -5, 5          # Batas untuk f1 dan f2
n = 10000

# Hasil untuk fungsi pertama
exact_f1 = exact_integral(f1, a1, b1)
rectangular_f1 = rectangular_integral(f1, a1, b1, n)
trapezoidal_f1 = trapezoidal_integral(f1, a1, b1, n)
monte_carlo_f1 = monte_carlo_integral(f1, a1, b1, n)

# Hasil untuk fungsi kedua
exact_f2 = exact_integral(f2, a1, b1)
rectangular_f2 = rectangular_integral(f2, a1, b1, n)
trapezoidal_f2 = trapezoidal_integral(f2, a1, b1, n)
monte_carlo_f2 = monte_carlo_integral(f2, a1, b1, n)

# Parameter untuk fungsi ketiga dan keempat
a3, b3 = 0, np.pi      # Batas untuk f3
a4, b4 = -2, 2         # Batas untuk f4

# Hasil untuk fungsi ketiga
exact_f3 = exact_integral(f3, a3, b3)
rectangular_f3 = rectangular_integral(f3, a3, b3, n)
trapezoidal_f3 = trapezoidal_integral(f3, a3, b3, n)
monte_carlo_f3 = monte_carlo_integral(f3, a3, b3, n)

# Hasil untuk fungsi keempat
exact_f4 = exact_integral(f4, a4, b4)
rectangular_f4 = rectangular_integral(f4, a4, b4, n)
trapezoidal_f4 = trapezoidal_integral(f4, a4, b4, n)
monte_carlo_f4 = monte_carlo_integral(f4, a4, b4, n)

# Print hasil
print("Fungsi 1: (1 / √(2π)) * exp(-x² / 2)")
print(f"Integral Eksak: {exact_f1:.4f}")
print(f"Integral Rectangular: {rectangular_f1:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f1:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f1:.4f}")

print("\nFungsi 2: 3x²")
print(f"Integral Eksak: {exact_f2:.4f}")
print(f"Integral Rectangular: {rectangular_f2:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f2:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f2:.4f}")

print("\nFungsi 3: sin(x)")
print(f"Integral Eksak: {exact_f3:.4f}")
print(f"Integral Rectangular: {rectangular_f3:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f3:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f3:.4f}")

print("\nFungsi 4: e^(-x²)")
print(f"Integral Eksak: {exact_f4:.4f}")
print(f"Integral Rectangular: {rectangular_f4:.4f}")
print(f"Integral Trapezoidal: {trapezoidal_f4:.4f}")
print(f"Integral Monte Carlo: {monte_carlo_f4:.4f}")

# Visualisasi fungsi
x1 = np.linspace(a1, b1, 100)
x2 = np.linspace(a1, b1, 100)
x3 = np.linspace(a3, b3, 100)
x4 = np.linspace(a4, b4, 100)

plt.figure(figsize=(12, 10))

# Subplot untuk fungsi pertama
plt.subplot(2, 2, 1)
plt.plot(x1, f1(x1), color='blue', label='(1 / √(2π)) * exp(-x² / 2)')
plt.fill_between(x1, f1(x1), color='lightblue', alpha=0.5)
plt.title('Fungsi 1: (1 / √(2π)) * exp(-x² / 2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Subplot untuk fungsi kedua
plt.subplot(2, 2, 2)
plt.plot(x2, f2(x2), color='orange', label='3x²')
plt.fill_between(x2, f2(x2), color='lightyellow', alpha=0.5)
plt.title('Fungsi 2: 3x²')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Subplot untuk fungsi ketiga
plt.subplot(2, 2, 3)
plt.plot(x3, f3(x3), color='green', label='sin(x)')
plt.fill_between(x3, f3(x3), color='lightgreen', alpha=0.5)
plt.title('Fungsi 3: sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Subplot untuk fungsi keempat
plt.subplot(2, 2, 4)
plt.plot(x4, f4(x4), color='red', label='e^(-x²)')
plt.fill_between(x4, f4(x4), color='lightpink', alpha=0.5)
plt.title('Fungsi 4: e^(-x²)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
